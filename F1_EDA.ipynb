*Meaning of Columns in lap*
<br>*Variables colored in Red will be used in Data Analysis (important columns I think)

1. **sessionUID** - ID of session (Unique)
2. **<span style = "color:Red">sessionTime</span>** -  session timestamp
3. **<span style = "color:Red">lastLapTimeInMS</span>** - last lap time in milliseconds
4. **<span style = "color:Red">currentLapTimeInMS</span>** - current time around the lap in milliseconds
5. **<span style = "color:Red">sector1TimeInMS</span>** - sector 1 time in milliseconds
6. **<span style = "color:Red">sector2TimeInMS</span>** - sector 2 time in milliseconds
7. **<span style = "color:Red">lapDistance</span>** - distance vehicle is around current lap in metres ('-' could)
            <br>be negative if line hasn't been crossed yet
8. **<span style = "color:Red">totalDistance</span>** - total distance travelled in session in metres ('-' could)
            <br>be negative if line hasn't been crossed yet
9. **safetyCarDelta** - delta in seconds for safety car
10. **carPosition** - car race position
11. **<span style = "color:Red">currentLapNum</span>** - current lap number
12. **<span style = "color:Red">pitStatus</span>** - 0 = none / 1 = pitting / 2 = in pit area
13. **<span style = "color:Red">numPitStops</span>** - number of pit stops taken in this race
14. **<span style = "color:Red">sector</span>** - 0 = sector1 / 1 = sector2 / 2 = sector3
15. **<span style = "color:Red">currentLapInvalid</span>** - current lap invalid (0 = valid / 1 = invalid)
16. **<span style = "color:Red">penal
    ties</span>** - accumulated time penalties in seconds to be added
17. **<span style = "color:Red">warnings</span>** - accumulated number of warnings issued
18. **numUnservedDriveThroughPens** - num drive through pens left to serve
19. **numUnservedStopGoPens** - num stop go pens left to serve
20. **gridPosition** - grid position the vehicle started the race in
21. **<span style = "color:Red">driverStatus</span>** - status of driver (0 = in garage / 1 = flying lap / 2 = in lap / 3 = out lap / 4 = on track)
22. **<span style = "color:Red">resultStatus</span>** - result status (0 = invalid / 1 = inactive / 2 = active / 3 = finished / 4 = didnotfinish / 5 = disqualified / 6 = not classified / 7 = retired)
23. **pitLaneTimerActive** - pit lane timing (0 = inactive / 1 = active)
24. **pitLaneTimeInLaneInMS** - if active, the current time spent in the pit lane in milliseconds
25. **pitStopTimerInMS** - time of the actual pit stop in milliseconds
26. **pitStopShouldServePen** - whether the car should serve a penalty at this stop

## motion dataframe
- Cotains all motion data for player's car (only sent while player is in control)
- Physics data for all the cars being driven
- There is additional data for the car being driven with the goal of being able to drive a motion platform setup
- **N.B.** For the normalised vectors below, to convert to float values divide by 32767.0f – 16-bit signed values are used to pack the data and on the assumption that direction values are always between -1.0f and 1.0f.

```python
motion.info()
```

## session dataframe
- Data about the session (track, time left)
- Details about the current session in progress

```python
session.info()
```

*Meaning of Columns in session*
1. **sessionUID** - ID of session (Unique)
2. <span style = "color:Red">**sessionTime**</span> -  session timestamp
3. **weather** - 0 = clear / 1 = light cloud / 2 = overcast / 3 = light rain / 4 = heavy rain / 5 = storm 
4. **trackTemperature** - track temp in degrees Celsius
5. **airTemperature** - air temp in degrees Celsius
6. <span style = "color:Red">**totalLaps**</span> - total number of laps in this race
7. <span style = "color:Red">**trackLength**</span> - total length in metres
8. <span style = "color:Red">**sessionType**</span> - 0 = unknown / 1 = P1 / 2 = P2 / 3 = P3 / 4 = Short / 5 = Q1 / 6 = Q2 / 7 = Q3 / 8 = Short Q / 9 = OSQ / 10 = R / 11 = R2 / 12 = R3 / 13 = Time Trial
9. <span style = "color:Red">**trackId**</span> - '-1' for unknown, see appendix
10. **formula** - 0 = F1 Modern / 1 = F1 Cclassic / 2 = F2 / 3 = F1 Generic / 4 = Beta / 5 = Supercars
11. <span style = "color:Red">**sessionTimeLeft**</span> - time left in session in seconds
12. <span style = "color:Red">**sessionDuration**</span> - session duration in seconds
13. **pitSpeedLimit** - pit speed limit in kilometers per hour
14. **gamePaused** - whether the game is paused (network game only)
15. **isSpectating** - whether the player is spectating
16. **spectatorCarIndex** - Index of the car being spectated
17. **sliProNativeSupport** - SLI Pro support (0 = inactive / 1 = active)
18. **numMarshalZones** - number of marshal zones to follow
19. **safetyCarStatus** - 0 = no safety car / 1 = full / 2 = virtual / 3 = formation lap
20. **networkGame** - 0 = offline / 1 = online
21. **forecastAccuracy** - 0 = Perfect / 1 = Approximate
22. **aiDifficulty** - AI difficulty rating (0 ~ 110).
    - Adjust to determine the skill level of AI competitors. This option has a value range of Very Easy (0-19), Easy (20-39), Medium (40-59), Hard(60-79), Expert(80-89), Master(90-94), Legend (95-100), and Ultimate (101-110).The default is set to 31 (Easy).
23. **seasonLinkIdentifier** - identifier for season (persists across saves)
24. **weekendLinkIdentifier** - identifier for weekend (persists across saves)
25. **sessionLinkIdentifier** - identifier for session (persists across saves)
26. **pitStopWindowIdealLap** - ideal lap to pit on for current strategy (player)
27. **pitStopWindowLatestLap** - latest lap to pit on for current strategy (player)
28. **pitStopRejoinPosition** - predicted position to rejoin at (player)
29. **steeringAssist** - 0 = off / 1 = on
30. **brakingAssist** - 0 = off / 1 = low / 2 = medium / 3 = high
31. **gearboxAssist** - 1 = manual / 2 = manual & suggested gear / 3 = auto
32. <span style = "color:Red">**pitAssist**</span> - 0 = off / 1 = on
33. **pitReleaseAssist** - 0 = off / 1 = on
34. <span style = "color:Red">**ERSAssist**</span> - 0 = off / 1 = on
35. <span style = "color:Red">**DRSAssist**</span> - 0 = off / 1 = on
36. **dynamicRacingLine** - 0 = off / 1 = corners only / 2 = full
37. **dynamicRacingLineType** - 0 = 2D / 1 = 3D
38. <span style = "color:Red">**gameMode**</span> - game mode id (see appendix)
39. <span style = "color:Red">**ruleSet**</span> - ruleset (see appendix)
40. **timeOfDay** - local time of day (minutes since midnight)
41. <span style = "color:Red">**sessionLength**</span> - 0 = None / 2 = Very Short / 3 = Short / 4 = Medium / 5 = Medium Long / 6 = Long / 7 = Full
